```python
class complex_number:
    def calculate_area(a, b, c):# Метод для расчета площади треугольника
        s = (a + b + c) / 2
        area = (s * (s - a) * (s - b) * (s - c)) ** 0.5 # Формула Герона для вычисления площади треугольника
        return area

    def calculate_perimeter(a, b, c): # Метод для расчета периметра треугольника
        perimeter = a + b + c
        return perimeter

a = float(input("Введите сторону a: "))
b = float(input("Введите сторону b: "))
c = float(input("Введите сторону c: "))

print("Площадь треугольника:", complex_number.calculate_area(a, b, c))
print("Периметр треугольника:", complex_number.calculate_perimeter(a, b, c))
```

```python
class Salary:
    def __init__(self, surname, name, patronymic, start_year_work, salary, bonus_percent, worked_days, total_work_days):
       
    # Присвоение атрибутов экземпляра класса значениями из параметров
        self.surname = surname #фамилия
        self.name = name #имя
        self.patronymic = patronymic #отчество
        self.start_year_work = start_year_work  # год поступления на работу
        self.salary = salary # оклад в руб
        self.bonus_percent = bonus_percent  # процент надбавки
        self.worked_days = worked_days  # кол-во отработанных дней
        self.total_work_days = total_work_days  # кол-во рабочих дней в месяце
        self.accrued_amount = 0  # начисленная сумма
        self.withheld_amount = 0 # удержанная сумма

    # Метод для расчета начисленной суммы зарплаты
    def calculate_accrued_amount(self):
        bonus = self.salary * (self.bonus_percent / 100)
        self.accrued_amount = (self.salary / self.total_work_days) * self.worked_days + bonus

    # Метод для расчета удержанной суммы
    def calculate_withheld_amount(self):
        self.withheld_amount = self.accrued_amount * 0.13

    # Метод для расчета чистой заработной платы
    def net_salary(self):
        return self.accrued_amount - self.withheld_amount

    # Метод для расчета стажа работы сотрудника
    def work_experience(self):
        current_year = 2024
        experience_years = current_year - self.start_year_work
        return experience_years   
    
surname = input("Введите фамилию сотрудника: ")
name = input("Введите имя сотрудника: ")
patronymic = input("Введите отчество сотрудника: ")
start_year_work = int(input("Введите год начала работы: "))
salary = float(input("Введите зарплату сотрудника: "))
bonus_percent = float(input("Введите процент премии: "))
worked_days = int(input("Введите количество отработанных дней: "))
total_work_days = int(input("Введите общее количество рабочих дней: "))


# Создание экземпляра класса Salary с введенными данными и расчетом зарплаты
employee = Salary(surname, name, patronymic, start_year_work, salary, bonus_percent, worked_days, total_work_days)
employee.calculate_accrued_amount()
employee.calculate_withheld_amount()


print("Начисленная сумма:", employee.accrued_amount)
print("Удержанная сумма:", employee.withheld_amount)
print("Сумма, выдаваемая на руки:", employee.net_salary())
print("Стаж работы:", employee.work_experience(), "лет/год")
```

```python
import random 

class Equation:
    def calculate_root(a, b, c):
        discriminant = b**2 - 4*a*c
        if discriminant > 0:
            root1 = (-b + discriminant**(0.5)) / (2*a)
            root2 = (-b - discriminant**(0.5)) / (2*a)
            return root1, root2
        elif discriminant == 0:
            root = -b / (2*a)
            return root
        else:
            return "Нет корней"

    def display_result(root):
        print(f"Корень(и) уравнения: {root}")

class LinearEquation(Equation):

    def calculate_root(a, b):
        root = -b / a
        return root

    def display_info(a, b):
        print(f"Линейное уравнение: {a}x + {b} = 0")

class QuadraticEquation(Equation):

    def calculate_root(a, b, c):
        discriminant = b**2 - 4 * a * c
        if discriminant > 0:
            root1 = (-b + discriminant**0.5) / (2 * a)
            root2 = (-b - discriminant**0.5) / (2 * a)
            return root1, root2
        elif discriminant == 0:
            root = -b / (2 * a)
            return root
        else:
            return "нет корней"

    def display_info(a, b, c):
        print(f"Квадратное уравнение: {a}x^2 + {b}x + {c} = 0")

# Количество уравнений для создания
n = int(input("Введите количество уравнений: "))

# Создаем список n уравнений с случайными параметрами
equations = []
for _ in range(n):
    equation_type = LinearEquation if random.randint(0, 1) == 0 else QuadraticEquation
    equation_params = tuple(random.randint(1, 10) for _ in range(3)) if equation_type == QuadraticEquation else tuple(random.randint(1, 10) for _ in range(2))
    equations.append((equation_type, equation_params))

# Выводим полную информацию об уравнениях
for eq_type, args in equations:
    eq_type.display_info(*args)
    roots = eq_type.calculate_root(*args)
    print("Корни уравнения:", roots)
    print()
```
