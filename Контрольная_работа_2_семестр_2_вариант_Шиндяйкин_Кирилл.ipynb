1. Напишите программу, которая выполняет добавление, удаление и вывод на экран элементов списка при помощи классов.

3.	Создаем объект данного класса.
4.	Используя созданный объект, вызываем соответствующий метод (в зависимости от выбора пользователя).

В программе реализовать выбор с помощью меню:

```python
# Решение 1.
class List:
    def __init__(self):
        self.lst = []

    def add_element(self, element):
        self.lst.append(element)

    def remove_element(self, element):
        if element in self.lst:
            self.lst.remove(element)

    def display_list(self):
        print(self.lst)

my_list = List()

while True:
    print("0. Выход")
    print("1. Добавить")
    print("2. Удалить")
    print("3. Вывести на экран")

    choice = int(input("Выберите одно из этих значений: "))

    if choice == 0:
        break
    elif choice == 1:
        element = input("Введите элемент для добавления: ")
        my_list.add_element(element)
    elif choice == 2:
        element = input("Введите элемент для удаления: ")
        my_list.remove_element(element)
    elif choice == 3:
        my_list.display_list()
    else:
        print("Неверный выбор.")
```

Создайте список, содержащий по 2 объекта каждого класса (People, Worker, Teacher). Для этого списка:

• выведите информацию о каждом человеке с помощью метода info;

```python
# Решение 2.
class People:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        return f"Name: {self.name}, Age: {self.age}"

class Worker(People):
    def __init__(self, name, age, post, salary):
        super().__init__(name, age)
        self.post = post
        self.salary = salary

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Post: {self.post}, Salary: {self.salary}"

class Teacher(Worker):
    def __init__(self, name, age, post, salary):
        super().__init__(name, age, post, salary)
        self.disciplines = []

    def add_discipline(self, discipline):
        self.disciplines.append(discipline)

    def delete_discipline(self, discipline):
        if discipline in self.disciplines:
            self.disciplines.remove(discipline)

    def __str__(self):
        return f"Name: {self.name}, Age: {self.age}, Post: {self.post}, Salary: {self.salary}, Disciplines: {', '.join(self.disciplines)}"


person1 = People("Марина", 25)
person2 = People("Василий", 35)

worker1 = Worker("Марк", 30, "Менеджер", 50000)
worker2 = Worker("Елена", 31, "Доставщик", 60000)

teacher1 = Teacher("Вероника", 23, "Профессор", 70000)
teacher1.add_discipline("Математика")
teacher1.add_discipline("Физика")
teacher2 = Teacher("Максим", 29, "Учитель", 45000)
teacher2.add_discipline("История")
teacher2.add_discipline("Литература")

# List of objects
people_list = [person1, person2, worker1, worker2, teacher1, teacher2]

# Printing information
for person in people_list:
    print(person.info())

print("Список фамилий тех, кто моложе 30 лет:")
for person in people_list:
    if person.age < 30:
        print(person.name)

teacher1.add_discipline("Химия")
teacher2.delete_discipline("Литература")

print(teacher1)
print(teacher2)
```

3.Создайте класс Заказ(Order), у которого есть свойства кодтовара(code), цена(price), количество(count) и методы _init и str.

Создайте 2 класса-потомка: Опт(Opt) и Розница(Retail). В этих классах создайте методы init, str.и сумма_заказа(summa), позволяющий узнать стоимость заказа. Для опта стоимость единицы товара составляет 95% от цены, а при покупке более 500 штук – 90% цены. В розницу стоимость единицы товара составляет 100% цены. Стоимость заказа равна произведению цены на количество.

Продемонстрируйте работу с классами, создав необходимые объекты и обратившись к их свойствам и методам

```python
# Решение 3.
class Order:
    def __init__(self, code, price, count):
        self.code = code
        self.price = price
        self.count = count

    def __str__(self):
        return f"Code: {self.code}, Price: {self.price}, Count: {self.count}"

class Opt(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def summa(self):
        if self.count > 500:
            return 0.9 * self.price * self.count
        else:
            return 0.95 * self.price * self.count

    def __str__(self):
        return super().__str__()

class Retail(Order):
    def __init__(self, code, price, count):
        super().__init__(code, price, count)

    def summa(self):
        return self.price * self.count

    def __str__(self):
        return super().__str__()


order1 = Opt("SKU001", 200, 400)
order2 = Retail("SKU002", 100, 100)
order3 = Opt("SKU003", 200, 600)


print(order1)
print(f"Total cost for Opt: {order1.summa()}")

print(order2)
print(f"Total cost for Retail: {order2.summa()}")
print(order3)
print(f"Total cost for Retail: {order3.summa()}")
```

4.Написать функцию-генератор my_func_2(lst), которая принимает объект, поддерживающий итерации с произвольным уровнем вложенности, и возвращает все элементы по одному.

```python
# Решение 4.
def my_func_2(lst):
    if isinstance(lst, list):
        for item in lst:
            if isinstance(item, list):
                yield from my_func_2(item)
            else:
                yield item
    else:
        yield lst


input_data = input("Введите данные в формате списка (например, [1, 2, [3, 4], 5]): ")
data = eval(input_data)  # Преобразование введенной строки в список
result = list(my_func_2(data))
print("Результат:", result)
```

Слова в тексте разделены пробелами. Написать реализацию в одну строку. Оформить решение в виде функции my_func_3(text), т.е. шаблон таков:

строка с import, если необходимо def my_func_3(text): return # однострочная реализация задания

```python
# Решение 5.


def my_func_3(text):
    return sum(map(lambda word: len(word) >= 3, text.split()))


text = input('Введи тест: ') 
print(my_func_3(text)) 
```
